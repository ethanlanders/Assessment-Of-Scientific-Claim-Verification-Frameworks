```python
import pandas as pd
import numpy as np

#1
df = pd.read_csv("https://www.cs.odu.edu/~sampath/courses/data/earthquakes.csv")
print("#1\n", df, "\n")

#2
print("#2\nShape:  ", df.shape)
print("Columns:  \n", df.columns, "\n")
print(df.head(), "\n")

#3
unique_values = df['alert'].unique()
unique_values = unique_values[1:]
print("#3\nUnique values available in the 'alert' column:  ", unique_values, '\n')
# other unique values in the 'alert' column include 'yellow' and 'orange
# according to the USGS API documentation

#4
print("#4\n", df[['title', 'time']].iloc[100:106], "\n")

#5
column_names = ['mag', 'title', 'parsed_place', 'time']
selected_columns = [col for col in df.columns if any(col.startswith(column_name) for column_name in column_names)]
print("#5\n", selected_columns, "\n")

#6
selected_columns_df = df[selected_columns]
condition = selected_columns_df["mag"]>=6
new_df = selected_columns_df[condition]
print("#6\n", new_df[['parsed_place', 'mag']], "\n")

#7
condition = (df['alert']=='red') & (df['tsunami']==1)
filtered_df = df[condition]
print("#7\n", filtered_df[['parsed_place', 'mag']], "\n")

#8
condition = (df['magType'] == 'mb')
new_df = df[condition]
print("#8\nThe 90th percentile of earthquake magnitude in Japan using the magType of 'mb':  ",
      new_df['mag'].quantile(0.9), "\n")

#9
condition = (df['type']=='earthquake') & (df['parsed_place']=='California')
filtered_df = df[condition]
print("#9\n", filtered_df.describe(), "\n")

#10
df['ring_of_fire'] = "no"
places = ['Bolivia', 'Chile', 'Ecuador', 'Peru', 'Costa Rica', 'Guatemala', 'Mexico',
          'Japan', 'Philippines', 'Indonesia', 'New Zealand', 'Antarctic', 'Canada',
          'Fiji', 'Alaska', 'Washington', 'California', 'Russia', 'Taiwan', 'Tonga',
          'Kermadec Islands']
condition = df["parsed_place"].isin(places)
df.loc[condition, "ring_of_fire"] = "yes"
print("#10\n", df[["parsed_place", "ring_of_fire"]])
```
